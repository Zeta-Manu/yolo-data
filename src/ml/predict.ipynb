```python
from ultralytics import YOLO

# Load a model
model = YOLO('runs/detect/train21/weights/last.pt')  # load a custom model
```

```python
source = 'https://www.youtube.com/shorts/mN65mYMhR0M'
threshold = 0.5
results = model.predict(source, stream=True, conf=threshold)
```

```python
result_arr = []
for result in results:
    result_arr.append(result)
```

```python
names = result_arr[0].names
print(names)
```

```python
for r in result_arr:
    data = r.boxes
    for box in data:
        print("Object type:", box.cls.item())
        print("Object name:", names[box.cls[0].item()])
        print("Coordinates:", box.xyxy.tolist())
        print("Probability:", box.conf.item())
```

```python
import json
# Example Inference Prediction
data = {
    "id": box.cls.item(),
    "name": names[box.cls[0].item()],
    "coordinate": box.xyxy.tolist(),
    "prob": box.conf.item(), 
}

# Convert data to JSON
json_data = json.dumps({"data": data}, indent=4)

# Print the JSON data
print(json_data)
```
